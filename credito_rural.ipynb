```python
# AQUI NÓS ESTAMOS PEGANDO OS ARQUIVOS SHPs DO CAR DE TODOS OS ESTADOS DA AMAZÔNIA
#OBJETIVO É OBTER A ÁREA DAS FAZENDAS FINANCIADAS PELO CREDITO RURAL 


import os
import zipfile
import geopandas as gpd
import pandas as pd

zip_dir = r"D:\Infoamazonia\shapes\CAR\CAR_ESTADOS"
extract_dir = r"D:\Infoamazonia\shapes\CAR\extracted"

# Certificar-se de que o diretório de extração existe
os.makedirs(extract_dir, exist_ok=True)

# Lista para armazenar os GeoDataFrames
gdframes = []

# Processar cada arquivo ZIP no diretório
for zip_file in os.listdir(zip_dir):
    if zip_file.endswith(".zip"):
        zip_path = os.path.join(zip_dir, zip_file)
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            # Extrair os arquivos para um subdiretório específico para cada ZIP
            subdir = os.path.join(extract_dir, os.path.splitext(zip_file)[0])
            os.makedirs(subdir, exist_ok=True)
            zip_ref.extractall(subdir)
            print(f"Arquivos extraídos de {zip_file} para {subdir}")

            # Encontrar os arquivos .shp no subdiretório
            shp_files = [os.path.join(subdir, file) for file in os.listdir(subdir) if file.endswith('.shp')]

            # Carregar cada shapefile em um GeoDataFrame
            for shp in shp_files:
                gdf = gpd.read_file(shp)
                gdframes.append(gdf)
                print(f"Carregado shapefile: {shp}")

# Concatenar todos os GeoDataFrames em um único GeoDataFrame
if gdframes:
    car = gpd.GeoDataFrame(pd.concat(gdframes, ignore_index=True))
    print("Todos os shapefiles foram concatenados com sucesso!")
    print(car.head())  # Exibir as primeiras linhas do GeoDataFrame combinado

    # Salvar o GeoDataFrame combinado em um novo shapefile
    output_shp = r"D:\Infoamazonia\shapes\CAR\combined_shapefile.shp"
    car.to_file(output_shp)
    print(f"GeoDataFrame combinado salvo em: {output_shp}")
else:
    print("Nenhum shapefile foi encontrado nos arquivos ZIP.")
```

```python
## GERANDO DF DAS ÁREAS FINANCIADAS, INCLUINDO O ID_BACEN NO MAPA ##


import geopandas as gpd
import pandas as pd
from shapely.validation import make_valid

# ---------------------------------------------------------
# 0) Tipos e pré-condições
# ---------------------------------------------------------
# Garanta que os IDs usados no join sejam strings
credito['id_referencia_bacen'] = credito['id_referencia_bacen'].astype('string')
proj['id_referencia_bacen'] = proj['id_referencia_bacen'].astype('string')

# ---------------------------------------------------------
# 1) Merge CAR x CREDITO (mantendo geometria do CAR)
# ---------------------------------------------------------
df_matched = car.merge(credito, left_on='cod_imovel', right_on='id_car', how='inner')

# Certifique-se de que continua sendo GeoDataFrame e com CRS do CAR
df_matched = gpd.GeoDataFrame(df_matched, geometry='geometry', crs=car.crs)

# Geometrias válidas
df_matched['geometry'] = df_matched['geometry'].apply(make_valid)

# ---------------------------------------------------------
# 2) Join com PROJ via id_referencia_bacen
# ---------------------------------------------------------
financiamento = df_matched.merge(proj, on='id_referencia_bacen', how='inner')
financiamento = gpd.GeoDataFrame(financiamento, geometry='geometry', crs=df_matched.crs)

# ---------------------------------------------------------
# 3) Agregação por imóvel (dissolve a geometria)
# ---------------------------------------------------------
def _uniq_join(series):
    # remove NaN, converte para string, pega valores únicos e ordena
    vals = series.dropna().astype('string').unique().tolist()
    return ', '.join(sorted(vals))

df_grouped = (
    financiamento
    .groupby('cod_imovel', as_index=False)
    .agg({
        'geometry': lambda s: s.unary_union,   # dissolve as geometrias do grupo
        'id_referencia_bacen': _uniq_join,
        'valor_parcela_credito': _uniq_join,
        'mod_fiscal': 'first',
        'num_area': 'first',
        'ind_status': 'first',
        'ind_tipo': 'first',
        'des_condic': 'first',
        'municipio': 'first',
        'cod_estado': 'first',
        'id_sncr': _uniq_join,
        'id_nirf': _uniq_join,
        'tipo_cpf_cnpj': _uniq_join,
        'data_emissao': _uniq_join,
        'data_vencimento': _uniq_join,
        'id_categoria_emitente': _uniq_join,
        'id_fase_ciclo_producao': _uniq_join,
        'ano': _uniq_join,
	    'id_tipo_cultivo': _uniq_join, 
	    'cnpj_basico_instituicao_financeira': _uniq_join, 
	    'id_car': _uniq_join
    })
)

# ---------------------------------------------------------
# 4) Volta a GeoDataFrame e reprojeta para WGS84 (EPSG:4326)
# ---------------------------------------------------------
gdf_grouped = gpd.GeoDataFrame(df_grouped, geometry='geometry', crs=financiamento.crs)
gdf_grouped = gdf_grouped.to_crs(4326)

# Validar novamente após reprojeção (evita geometrias inválidas pós-transformação)
gdf_grouped['geometry'] = gdf_grouped['geometry'].apply(make_valid)

# ---------------------------------------------------------
# 5) Exporta para GeoPackage
# ---------------------------------------------------------
gdf_grouped.to_file(
    r'D:\Python Scripts\BANCOS\mapa_credito_rural\propriedades_base_1.gpkg',
    layer='imoveis_agrupados',
    driver='GPKG'
)

print(gdf_grouped.head())
```

```python
import geopandas as gpd
import pandas as pd

p_grouped     = r'D:\Python Scripts\BANCOS\mapa_credito_rural\propriedades_base_1.gpkg'
layer_grouped = 'imoveis_agrupados'   # ajuste se seu layer tiver outro nome
p_embargo_shp = r'D:/Python Scripts/BANCOS/mapa_credito_rural/embargos_CR.shp'

# --- ler dados ---
gdf_grouped = gpd.read_file(p_grouped, layer=layer_grouped)
gdf_embargo = gpd.read_file(p_embargo_shp)

# --- escolher colunas existentes para descrição/nome ---
col_desc = 'des_tad' if 'des_tad' in gdf_embargo.columns else ('desc_tad' if 'desc_tad' in gdf_embargo.columns else None)
col_nome = 'nome_embar' if 'nome_embar' in gdf_embargo.columns else ('nome' if 'nome' in gdf_embargo.columns else None)

# --- selecionar só o necessário ---
cols_needed = ['cod_imovel', 'area_calc']
if col_desc: cols_needed.append(col_desc)
if col_nome: cols_needed.append(col_nome)

df_emb = gdf_embargo[[c for c in cols_needed if c in gdf_embargo.columns]].copy()

if 'cod_imovel' not in df_emb.columns:
    raise KeyError("A camada de embargos não possui a coluna 'cod_imovel'.")

# --- tipos / numéricos ---
df_emb['cod_imovel']       = df_emb['cod_imovel'].astype(str)
gdf_grouped['cod_imovel']  = gdf_grouped['cod_imovel'].astype(str)
if 'area_calc' in df_emb.columns:
    df_emb['area_calc'] = pd.to_numeric(df_emb['area_calc'], errors='coerce')

# --- helper: lista única em texto (QGIS-friendly) ---
def _uniq_join_str(s: pd.Series) -> str:
    vals = [str(x).strip() for x in s.dropna().tolist() if str(x).strip() != '']
    return '; '.join(sorted(set(vals)))

# --- agregações por cod_imovel ---
if len(df_emb) == 0:
    embargos_por_car = pd.DataFrame({'cod_imovel': gdf_grouped['cod_imovel'].unique(),
                                     'embargo_count': 0})
else:
    # contagem total
    df_count = (df_emb.groupby('cod_imovel', as_index=False)
                      .size()
                      .rename(columns={'size': 'embargo_count'}))

    agg_kwargs = {}
    if col_desc:
        agg_kwargs['emb_des_tad'] = (col_desc, _uniq_join_str)  # lista em texto
    if col_nome:
        agg_kwargs['emb_nome'] = (col_nome, _uniq_join_str)     # lista em texto
    if 'area_calc' in df_emb.columns:
        agg_kwargs['embargo_area_calc_sum'] = ('area_calc', 'sum')

    df_agg = df_emb.groupby('cod_imovel', as_index=False).agg(**agg_kwargs) if agg_kwargs else df_count.copy()
    embargos_por_car = df_count.merge(df_agg, on='cod_imovel', how='left')

# --- merge no gdf_grouped (mesmo nº de linhas) ---
gdf_out = gdf_grouped.merge(embargos_por_car, on='cod_imovel', how='left')

# --- preencher nulos e derivados ---
gdf_out['embargo_count'] = gdf_out['embargo_count'].fillna(0).astype(int)

for c in ['emb_des_tad', 'emb_nome']:
    if c in gdf_out.columns:
        gdf_out[c] = gdf_out[c].fillna('')

if 'embargo_area_calc_sum' in gdf_out.columns:
    gdf_out['embargo_area_calc_sum'] = gdf_out['embargo_area_calc_sum'].fillna(0.0)
    gdf_out['embargo_area_ha'] = gdf_out['embargo_area_calc_sum'] / 10_000

# --- checagem e visualização ---
assert len(gdf_out) == len(gdf_grouped), "O merge alterou a contagem de linhas."

show_cols = ['cod_imovel','embargo_count','emb_des_tad','emb_nome','embargo_area_calc_sum','embargo_area_ha']
print(gdf_out[[c for c in show_cols if c in gdf_out.columns]].head(10))
```

```python
#carregando camada desmatamento gerada no QGIS 

p_desmat = r'D:\Python Scripts\BANCOS\mapa_credito_rural\desmatamento_CR.shp'

# =========================================
# 1) LEITURA ROBUSTA (com fallback de encoding)
# =========================================
def read_vector_with_encoding(path):
    # tenta normal
    try:
        return gpd.read_file(path)  # engine padrão (pyogrio)
    except UnicodeDecodeError:
        pass

    # tenta encodings comuns via pyogrio
    for enc in ['latin1', 'cp1252', 'iso-8859-1', 'utf-8-sig']:
        try:
            gdf = gpd.read_file(path, encoding=enc)
            print(f"✅ Lido com encoding='{enc}' via pyogrio")
            return gdf
        except Exception:
            continue

    # fallback para Fiona (permite forçar encoding via Env)
    try:
        import fiona
        for enc in ['latin1', 'cp1252', 'iso-8859-1']:
            try:
                with fiona.Env(encoding=enc):
                    gdf = gpd.read_file(path, engine='fiona')
                    print(f"✅ Lido com encoding='{enc}' via fiona")
                    return gdf
            except Exception:
                continue
    except Exception:
        pass

    raise RuntimeError("Não foi possível ler o shapefile com os encodings testados.")

gdf_desm = read_vector_with_encoding(p_desmat)

# =========================================
# 2) CHECAGEM E NORMALIZAÇÃO DE COLUNAS
# =========================================
cols_expected = ['id_car', 'area_km', 'year', 'uuid']
missing = [c for c in cols_expected if c not in gdf_desm.columns]
if missing:
    raise KeyError(f"Colunas faltando em desmatamento: {missing}. Ajuste os nomes no código.")

gdf_desm['id_car'] = gdf_desm['id_car'].astype(str).str.strip()
gdf_out['id_car']  = gdf_out['id_car'].astype(str)

gdf_desm['area_km'] = pd.to_numeric(gdf_desm['area_km'], errors='coerce')   # km²
gdf_desm['year']    = pd.to_numeric(gdf_desm['year'], errors='coerce', downcast='integer')

# =========================================
# 3) AGREGAÇÃO POR id_car
# =========================================
def _uniq_join_str(series, sep=', '):
    vals = [str(x).strip() for x in series.dropna().tolist() if str(x).strip() != '']
    return sep.join(sorted(set(vals)))

desm_agg = (
    gdf_desm
      .groupby('id_car', as_index=False)
      .agg(
          desm_area_km_sum=('area_km', 'sum'),
          desm_anos_str=('year', lambda s: _uniq_join_str(s.astype('Int64').astype(str))),
          desm_uuid_str=('uuid', _uniq_join_str)
      )
)

# km² -> hectares
desm_agg['Desmatamento(ha)'] = (desm_agg['desm_area_km_sum'].fillna(0.0) * 100.0).round(2)

# =========================================
# 4) CASO id_car NO gdf_out ESTEJA EM LISTA DE TEXTO (explode → merge → re-agrega)
# =========================================
car_map = (
    gdf_out[['id_car']].copy()
      .assign(_row_id=lambda df: df.index)   # preserva índice original
)

# separa por vírgula / ponto-e-vírgula
car_map['id_car'] = (
    car_map['id_car']
      .astype(str)
      .str.split(r'[,;]\s*', regex=True)
)

car_map = car_map.explode('id_car', ignore_index=False)
car_map['id_car'] = car_map['id_car'].astype(str).str.strip()
car_map = car_map[car_map['id_car'].notna() & (car_map['id_car'] != '')]

# merge explode com o agregado por CAR
merged = car_map.merge(
    desm_agg[['id_car','Desmatamento(ha)','desm_anos_str','desm_uuid_str']],
    on='id_car', how='left'
)

# re-agrega por linha original (_row_id)
def _sum(series):
    return float(np.nansum(pd.to_numeric(series, errors='coerce'))) if len(series) else 0.0

per_row = (
    merged
      .groupby('_row_id', as_index=False)
      .agg(
          **{
              'Desmatamento(ha)': ('Desmatamento(ha)', _sum),
              'Anos_com_desmatamento': ('desm_anos_str', _uniq_join_str),
              'uuid_desmatamento': ('desm_uuid_str', _uniq_join_str)
          }
      )
)

# =========================================
# 5) JUNTAR NO gdf_out (MESMO Nº DE LINHAS)
# =========================================
gdf_out = gdf_out.join(per_row.set_index('_row_id'), how='left')

# Defaults
gdf_out['Desmatamento(ha)'] = gdf_out['Desmatamento(ha)'].fillna(0.0).round(2)
for c in ['Anos_com_desmatamento', 'uuid_desmatamento']:
    gdf_out[c] = gdf_out[c].fillna('')

# Checagem simples
print(gdf_out.head(10))


```

```python
import pandas as pd
import numpy as np


usecols = ['cod_imovel', 'id_foco', 'ano']  # ajuste se o nome for diferente
df_queimadas = pd.read_csv(
    r'D:\Python Scripts\BANCOS\mapa_credito_rural\queimadas\2023-2023.csv',
    usecols=usecols,
    dtype={'cod_imovel': 'string', 'id_foco': 'string', 'ano': 'string'}
)
df_queimadas['ano'] = pd.to_numeric(df_queimadas['ano'], errors='coerce').astype('Int64')

def _uniq_join_str(series, sep=', '):
    vals = [str(x).strip() for x in series.dropna().tolist() if str(x).strip() != '']
    return sep.join(sorted(set(vals), key=lambda x: (len(x), x)))  # ordena por tamanho/valor

def focos_por_car(df_queimadas: pd.DataFrame) -> pd.DataFrame:
    """
    Recebe um DataFrame com colunas: 'cod_imovel' (CAR), 'id_foco', 'ano'.
    Retorna um DataFrame por CAR com:
      - 'focos_count' (número de id_foco únicos)
      - 'focos_anos_str' (anos únicos como string separada por vírgulas)
    """
    # ---------- validação mínima ----------
    required = {'cod_imovel', 'id_foco', 'ano'}
    missing = required - set(df_queimadas.columns)
    if missing:
        raise KeyError(f"Faltam colunas em df_queimadas: {sorted(missing)}")

    df = df_queimadas.copy()

    # ---------- normalização ----------
    df['cod_imovel'] = df['cod_imovel'].astype(str).str.strip()
    df['id_foco'] = df['id_foco'].astype(str).str.strip()

    # ano → inteiro (quando possível)
    df['ano'] = pd.to_numeric(df['ano'], errors='coerce').astype('Int64')

    # opcional: remover linhas sem chave mínima
    df = df.dropna(subset=['cod_imovel', 'id_foco'])

    # se existirem duplicatas de (cod_imovel, id_foco, ano), removê-las
    df = df.drop_duplicates(subset=['cod_imovel', 'id_foco', 'ano'])

    # ---------- agregação por CAR ----------
    out = (
        df.groupby('cod_imovel', as_index=False)
          .agg(
              focos_count=('id_foco', pd.Series.nunique),  # conta id_foco únicos
              focos_anos_str=('ano', lambda s: _uniq_join_str(s.dropna().astype(int).astype(str)))
          )
    )

    # garantir tipos
    out['focos_count'] = out['focos_count'].fillna(0).astype(int)
    out['focos_anos_str'] = out['focos_anos_str'].fillna('')

    return out


df_focos = focos_por_car(df_queimadas)




```

```python
import pandas as pd
import geopandas as gpd
import numpy as np

# ----------------------------------------
# Helpers
# ----------------------------------------
def _uniq_join_str(series, sep=', '):
    vals = [str(x).strip() for x in series.dropna().tolist() if str(x).strip() not in ('', 'nan', 'None')]
    return sep.join(sorted(set(vals)))

# ----------------------------------------
# Pré-condições (tipos)
# ----------------------------------------

gdf_out = gdf
gdf_out['cod_imovel'] = gdf_out['cod_imovel'].astype(str).str.strip()
base_car_cr_uc['cod_imovel'] = base_car_cr_uc['cod_imovel'].astype(str).str.strip()
base_car_cr_ti['cod_imovel'] = base_car_cr_ti['cod_imovel'].astype(str).str.strip()

# ----------------------------------------
# 1) Flags UC / TI
# ----------------------------------------
uc_flag = (
    base_car_cr_uc.groupby('cod_imovel', as_index=False)
    .size()
    .rename(columns={'size': 'uc_count'})
)
uc_flag['UC'] = np.where(uc_flag['uc_count'] > 0, 'sim', 'não')
uc_flag = uc_flag[['cod_imovel', 'UC']]

ti_flag = (
    base_car_cr_ti.groupby('cod_imovel', as_index=False)
    .size()
    .rename(columns={'size': 'ti_count'})
)
ti_flag['TI'] = np.where(ti_flag['ti_count'] > 0, 'sim', 'não')
ti_flag = ti_flag[['cod_imovel', 'TI']]

# ----------------------------------------
# 2) Nomes combinados (UC + TI) → "área_afetada"
# ----------------------------------------
# Garante colunas de nome
for df, col in [(base_car_cr_uc, 'nome'), (base_car_cr_ti, 'terrai_nom')]:
    if col not in df.columns:
        df[col] = ''

names_long = pd.concat(
    [
        base_car_cr_uc[['cod_imovel', 'nome']].rename(columns={'nome': 'nome_area'}),
        base_car_cr_ti[['cod_imovel', 'terrai_nom']].rename(columns={'terrai_nom': 'nome_area'})
    ],
    ignore_index=True
)

area_names = (
    names_long.groupby('cod_imovel', as_index=False)
    .agg(área_afetada=('nome_area', _uniq_join_str))
)

# ----------------------------------------
# 3) Mesclar no gdf_out (mantendo nº de linhas)
# ----------------------------------------
gdf_out = gdf_out.merge(uc_flag, on='cod_imovel', how='left')
gdf_out = gdf_out.merge(ti_flag, on='cod_imovel', how='left')
gdf_out = gdf_out.merge(area_names, on='cod_imovel', how='left')

# Defaults
gdf_out['UC'] = gdf_out['UC'].fillna('não')
gdf_out['TI'] = gdf_out['TI'].fillna('não')
gdf_out['área_afetada'] = gdf_out['área_afetada'].fillna('')

# Conferência rápida
print(gdf_out[['cod_imovel', 'UC', 'TI', 'área_afetada']].head(10))

# (Opcional) salvar em nova layer do GPKG
gdf_out.to_file(
    r'D:\Python Scripts\BANCOS\mapa_credito_rural\propriedades_base_completo.gpkg',
    layer='imoveis_agrupados_uc_ti',
    driver='GPKG'
)
```

```python
import geopandas as gpd

# 1) Selecionar as colunas desejadas
gdf_colunas = gdf[[
    'cod_imovel',
    'id_referencia_bacen',
    'valor_parcela_credito',
    'municipio',
    'ano',
    'tipo_cpf_cnpj',
    'cnpj_basico_instituicao_financeira',
    'embargo_count',
    'emb_des_tad',
    'emb_nome',
    'embargo_area_calc_sum',
    'total_CR',
    'Desmatamento(ha)',
    'Anos_com_desmatamento',
    'geometry',
    'focos_count',
    'focos_anos_str',
    'UC',
    'área_afetada',
    'TI'
]].copy()

# 2) Renomear colunas "problemáticas" para nomes seguros no Mapbox / Tippecanoe
rename_map = {
    'Desmatamento(ha)': 'Desmatamento_ha',
    'Anos_com_desmatamento': 'Anos_com_desmatamento',  # já está ok
    'área_afetada': 'area_afetada',                    # tirar acento
    'valor_parcela_credito': 'valor_parcela_credito',  # já ok
    'cnpj_basico_instituicao_financeira': 'cnpj_instituicao',  # ok
    'embargo_area_calc_sum': 'embargo_area_ha',        # nome mais amigável pra popup
    'tipo_cpf_cnpj': 'cpf_cnpj',
    'focos_anos_str': 'focos_anos',
    'focos_count': 'focos_count',                      # ok
    'emb_des_tad': 'emb_des_tad',                      # ok
    'emb_nome': 'emb_nome',                            # ok
    'UC': 'UC',                                        # ok
    'TI': 'TI',                                        # ok
    'total_CR': 'total_CR',                            # ok
    'id_referencia_bacen': 'id_referencia_bacen',  # ok
}

gdf_colunas_limpo = gdf_colunas.rename(columns=rename_map).copy()

# 3) Garantir CRS WGS84 antes de exportar (Mapbox exige EPSG:4326)
if gdf_colunas_limpo.crs is None or gdf_colunas_limpo.crs.to_epsg() != 4326:
    gdf_colunas_limpo = gdf_colunas_limpo.to_crs(4326)

# 4) Conferir resultado
print(gdf_colunas_limpo.columns)
print(gdf_colunas_limpo.head())
```
